```
import pandas as pd
import numpy as np
```

```
train = pd.read_csv('train.csv',index_col=0)
test = pd.read_csv('test.csv',index_col= 0)
submit = pd.read_csv('sample_submission.csv')
train[:5]
```

```
test = test.astype(int)
```

```
test[:5]
```

```
x_train = train.loc[:,'0':'63']
y_train = train.loc[:,'label']

x_test = test.loc[:,'0':'63']
```

```
from sklearn.preprocessing import StandardScaler, Normalizer
std = StandardScaler()
nor = Normalizer()

x_train_std = std.fit_transform(x_train)
x_test_std = std.fit_transform(x_test)
x_train_nor = nor.fit_transform(x_train)
x_test_nor = nor.fit_transform(x_test)
```

```
from collections import Counter
print(Counter(y_train))
```

```
submit[:5]
```

```
#KNN k = 7 , p = 2 weight = 'distance / 0.98888
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=7,p=2,weights='distance')
KNN.fit(x_train,y_train)

test_pred = KNN.predict(x_test)

for i in range(len(test_pred)):
  submit['label'][i] = test_pred[i]

submit.to_csv('submission.csv',header=True,index = None)
!kaggle competitions submit -c 2020-ml-p3 -f submission.csv -m "KNN"
```
