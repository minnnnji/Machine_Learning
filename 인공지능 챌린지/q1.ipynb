```
import pandas as pd
import numpy as np
```

```
train = pd.read_csv('train.csv',header = None, skiprows=1)
test = pd.read_csv('test.csv', header=None, skiprows=1)
submit = pd.read_csv('sample_submission.csv')

train[:5]
```

```
train.drop(0,axis=1,inplace=True)
train[:5]
```

```
test[:5]
```

```
x_train = train.loc[:,1:784]
y_train = train.loc[:,785]

x_test = test.loc[:,1:784]
```

```
from collections import Counter
print(Counter(y_train))
```

```
for i in range(len(train)):
  if y_train[i] == 'missing' or y_train[i] =='NAN':
    train.drop(i,axis = 0,inplace= True)
  
train.describe()
```

```
from imblearn.over_sampling import SMOTE
sm = SMOTE()
```

```
x_train_syn, y_train_syn = sm.fit_resample(x_train,y_train)
```

```
print(Counter(y_train_syn))
```

```
y_train_syn = y_train_syn.astype(int)
```

```
print(y_train_syn.dtype)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators = 10, learning_rate=0.1, max_depth= 5) # 0.72

xgb.fit(x_train_syn,y_train_syn)
```

```
x_train_syn[:5]
```

```
x_test = np.array(x_test)
x_test
```

```
x_test = test.loc[:,1:784]
x_test[:5]
```

```
test_pred = xgb.predict(x_test)

for i in range(len(test_pred)):
  submit['label'][i] = test_pred[i]

submit[:5]
```

```
submit.to_csv('submission.csv',header=True,index = None)
!kaggle competitions submit -c sejong-ai-challenge-p1 -f submission.csv -m "xgb"
```
